```python
%matplotlib inline
import numpy, scipy, pandas, matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
#import stanford_mir; stanford_mir.init() 
```

```python
x, sr = librosa.load('test2.wav')
ipd.Audio(x, rate=sr)
```

```python
plt.figure(figsize=(16, 5))
librosa.display.waveplot(x, sr, alpha=0.8)
```

```python
ipd.Image("https://upload.wikimedia.org/wikipedia/commons/thumb/e/ea/ADSR_parameter.svg/640px-ADSR_parameter.svg.png")
```

```python
#synthesizing a pure tone of 
T = 4.0 # seconds
f0 = 1047
sr = 22050
t = numpy.linspace(0, T, int(T*sr), endpoint=False) # time variable
x = 0.1*numpy.sin(1*numpy.pi*f0*t)
ipd.Audio(x, rate=sr)
```

```python
#the spectrum of the pure tone:
X = scipy.fft(x[:1047])
X_mag = numpy.absolute(X)        # spectral magnitude
f = numpy.linspace(0, sr, 1047)  # frequency variable
plt.figure(figsize=(14, 5))
plt.plot(f[:1047], X_mag[:1047]) # magnitude spectrum
plt.xlabel('Frequency (Hz)')
```
